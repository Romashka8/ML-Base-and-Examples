–∑–¥–µ—Å—å $K$ ‚Äî —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏, $C$ ‚Äî –∏—Å—Ç–∏–Ω–Ω–æ–µ —Ä–∞–∑–±–∏–µ–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏ –Ω–∞ –∫–ª–∞—Å—Å—ã. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, $h$ –∏–∑–º–µ—Ä—è–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –∫–∞–∂–¥—ã–π –∫–ª–∞—Å—Ç–µ—Ä —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –æ–±—ä–µ–∫—Ç–æ–≤ –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞, –∞ $c$ ‚Äî –Ω–∞—Å–∫–æ–ª—å–∫–æ –æ–±—ä–µ–∫—Ç—ã –æ–¥–Ω–æ–≥–æ –∫–ª–∞—Å—Å–∞ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –æ–¥–Ω–æ–º—É –∫–ª–∞—Å—Ç–µ—Ä—É. –≠—Ç–∏ –º–µ—Ä—ã –Ω–µ —è–≤–ª—è—é—Ç—Å—è —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–º–∏. –û–±–µ –≤–µ–ª–∏—á–∏–Ω—ã –ø—Ä–∏–Ω–∏–º–∞—é—Ç –∑–Ω–∞—á–µ–Ω–∏—è –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ $[0, 1]$, –∏ –±–æ–ª—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –±–æ–ª–µ–µ —Ç–æ—á–Ω–æ–π –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏. –≠—Ç–∏ –º–µ—Ä—ã –Ω–µ —è–≤–ª—è—é—Ç—Å—è –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–º–∏, –∫–∞–∫ $\text{ARI}$ –∏–ª–∏ $\text{AMI}$, –∏ –ø–æ—ç—Ç–æ–º—É –∑–∞–≤–∏—Å—è—Ç –æ—Ç —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤. –°–ª—É—á–∞–π–Ω–∞—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è –Ω–µ –±—É–¥–µ—Ç –¥–∞–≤–∞—Ç—å –Ω—É–ª–µ–≤—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø—Ä–∏ –±–æ–ª—å—à–æ–º —á–∏—Å–ª–µ –∫–ª–∞—Å—Å–æ–≤ –∏ –º–∞–ª–æ–º —á–∏—Å–ª–µ –æ–±—ä–µ–∫—Ç–æ–≤. –í —ç—Ç–∏—Ö —Å–ª—É—á–∞—è—Ö –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω–µ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å $\text{ARI}$. –û–¥–Ω–∞–∫–æ –ø—Ä–∏ —á–∏—Å–ª–µ –æ–±—ä–µ–∫—Ç–æ–≤ –±–æ–ª–µ–µ 1000 –∏ —á–∏—Å–ª–µ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–µ–Ω–µ–µ 10 –¥–∞–Ω–Ω–∞—è –ø—Ä–æ–±–ª–µ–º–∞ –Ω–µ —Ç–∞–∫ —è–≤–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∞ –∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–æ–∏–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞–Ω–∞.
–î–ª—è —É—á—ë—Ç–∞ –æ–±–µ–∏—Ö –≤–µ–ª–∏—á–∏–Ω $h$ –∏ $c$ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –≤–≤–æ–¥–∏—Ç—Å—è $V$-–º–µ—Ä–∞, –∫–∞–∫ –∏—Ö —Å—Ä–µ–¥–Ω–µ–µ –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–æ–µ:<br>
$V=2\cdot \frac{hc}{h + c}$<br>
–û–Ω–∞ —è–≤–ª—è–µ—Ç—Å—è —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–π –∏ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –¥–≤–µ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —Å—Ö–æ–∂–∏ –º–µ–∂–¥—É —Å–æ–±–æ–π.

‚úÖ<b>–°–∏–ª—É—ç—Ç</b> - –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –æ–ø–∏—Å–∞–Ω–Ω—ã—Ö –≤—ã—à–µ –º–µ—Ç—Ä–∏–∫, –¥–∞–Ω–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –Ω–µ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç –∑–Ω–∞–Ω–∏—è –∏—Å—Ç–∏–Ω–Ω—ã—Ö –º–µ—Ç–æ–∫ –æ–±—ä–µ–∫—Ç–æ–≤, –∏ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏, –∏—Å–ø–æ–ª—å–∑—É—è —Ç–æ–ª—å–∫–æ —Å–∞–º—É (–Ω–µ—Ä–∞–∑–º–µ—á–µ–Ω–Ω—É—é) –≤—ã–±–æ—Ä–∫—É –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏. –°–Ω–∞—á–∞–ª–∞ —Å–∏–ª—É—ç—Ç –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –æ—Ç–¥–µ–ª—å–Ω–æ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞. –û–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ $a$ ‚Äî —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç –¥–∞–Ω–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –¥–æ –æ–±—ä–µ–∫—Ç–æ–≤ –∏–∑ —Ç–æ–≥–æ –∂–µ –∫–ª–∞—Å—Ç–µ—Ä–∞, —á–µ—Ä–µ–∑ $b$ ‚Äî —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –æ—Ç –¥–∞–Ω–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –¥–æ –æ–±—ä–µ–∫—Ç–æ–≤ –∏–∑ –±–ª–∏–∂–∞–π—à–µ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞ (–æ—Ç–ª–∏—á–Ω–æ–≥–æ –æ—Ç —Ç–æ–≥–æ, –≤ –∫–æ—Ç–æ—Ä–æ–º –ª–µ–∂–∏—Ç —Å–∞–º –æ–±—ä–µ–∫—Ç). –¢–æ–≥–¥–∞ —Å–∏–ª—É—ç—Ç–æ–º –¥–∞–Ω–Ω–æ–≥–æ –æ–±—ä–µ–∫—Ç–∞ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –≤–µ–ª–∏—á–∏–Ω–∞:<br>
$s = \frac{b - a}{\max(a, b)}$<br>
–°–∏–ª—É—ç—Ç–æ–º –≤—ã–±–æ—Ä–∫–∏ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è —Å—Ä–µ–¥–Ω—è—è –≤–µ–ª–∏—á–∏–Ω–∞ —Å–∏–ª—É—ç—Ç–∞ –æ–±—ä–µ–∫—Ç–æ–≤ –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å–∏–ª—É—ç—Ç –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ –æ–±—ä–µ–∫—Ç–æ–≤ —Å–≤–æ–µ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –¥–æ –æ–±—ä–µ–∫—Ç–æ–≤ –¥—Ä—É–≥–∏—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤. –î–∞–Ω–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ –ª–µ–∂–∏—Ç –≤ –¥–∏–∞–ø–∞–∑–æ–Ω–µ $[-1, 1]$. –ó–Ω–∞—á–µ–Ω–∏—è, –±–ª–∏–∑–∫–∏–µ –∫ -1, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –ø–ª–æ—Ö–∏–º (—Ä–∞–∑—Ä–æ–∑–Ω–µ–Ω–Ω—ã–º) –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è–º, –∑–Ω–∞—á–µ–Ω–∏—è, –±–ª–∏–∑–∫–∏–µ –∫ –Ω—É–ª—é, –≥–æ–≤–æ—Ä—è—Ç –æ —Ç–æ–º, —á—Ç–æ –∫–ª–∞—Å—Ç–µ—Ä—ã –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è –∏ –Ω–∞–∫–ª–∞–¥—ã–≤–∞—é—Ç—Å—è –¥—Ä—É–≥ –Ω–∞ –¥—Ä—É–≥–∞, –∑–Ω–∞—á–µ–Ω–∏—è, –±–ª–∏–∑–∫–∏–µ –∫ 1, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç "–ø–ª–æ—Ç–Ω—ã–º" —á–µ—Ç–∫–æ –≤—ã–¥–µ–ª–µ–Ω–Ω—ã–º –∫–ª–∞—Å—Ç–µ—Ä–∞–º. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —á–µ–º –±–æ–ª—å—à–µ —Å–∏–ª—É—ç—Ç, —Ç–µ–º –±–æ–ª–µ–µ —á–µ—Ç–∫–æ –≤—ã–¥–µ–ª–µ–Ω—ã –∫–ª–∞—Å—Ç–µ—Ä—ã, –∏ –æ–Ω–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç —Å–æ–±–æ–π –∫–æ–º–ø–∞–∫—Ç–Ω—ã–µ, –ø–ª–æ—Ç–Ω–æ —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –æ–±–ª–∞–∫–∞ —Ç–æ—á–µ–∫.<br>
–° –ø–æ–º–æ—â—å—é —Å–∏–ª—É—ç—Ç–∞ –º–æ–∂–Ω–æ –≤—ã–±–∏—Ä–∞—Ç—å –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ $k$ (–µ—Å–ª–∏ –æ–Ω–æ –∑–∞—Ä–∞–Ω–µ–µ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ) ‚Äî –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –º–∞–∫—Å–∏–º–∏–∑–∏—Ä—É—é—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–∏–ª—É—ç—Ç–∞. –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –º–µ—Ç—Ä–∏–∫, —Å–∏–ª—É—ç—Ç –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ñ–æ—Ä–º—ã –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –∏ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç –±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–∞ –±–æ–ª–µ–µ –≤—ã–ø—É–∫–ª—ã—Ö –∫–ª–∞—Å—Ç–µ—Ä–∞—Ö, –ø–æ–ª—É—á–∞–µ–º—ã—Ö —Å –ø–æ–º–æ—â—å—é –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤, –æ—Å–Ω–æ–≤–∞–Ω–Ω—ã—Ö –Ω–∞ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style='white')
%matplotlib inline
from sklearn import decomposition
from sklearn import datasets
from mpl_toolkits.mplot3d import Axes3D
```

```python
# –ù–∞—á–Ω—ë–º —Å —Ç–æ–≥–æ, —á—Ç–æ –Ω–∞—Å—ã–ø–µ–º –Ω–∞ –ø–ª–æ—Å–∫–æ—Å—Ç—å —Ç—Ä–∏ –∫–ª–∞—Å—Ç–µ—Ä–∞ —Ç–æ—á–µ–∫
X = np.zeros((150, 2))

np.random.seed(seed=42)
X[:50, 0] = np.random.normal(loc=0.0, scale=.3, size=50)
X[:50, 1] = np.random.normal(loc=0.0, scale=.3, size=50)

X[50:100, 0] = np.random.normal(loc=2.0, scale=.5, size=50)
X[50:100, 1] = np.random.normal(loc=-1.0, scale=.2, size=50)

X[100:150, 0] = np.random.normal(loc=-1.0, scale=.2, size=50)
X[100:150, 1] = np.random.normal(loc=2.0, scale=.5, size=50)

plt.figure(figsize=(5, 5))
plt.plot(X[:, 0], X[:, 1], 'bo');
```

```python
# –í scipy –µ—Å—Ç—å –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è —Å—á–∏—Ç–∞–µ—Ç —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è
# –º–µ–∂–¥—É –ø–∞—Ä–∞–º–∏ —Ç–æ—á–µ–∫ –∏–∑ –¥–≤—É—Ö –º–∞—Å—Å–∏–≤–æ–≤, –ø–æ–¥–∞—é—â–∏—Ö—Å—è –µ–π –Ω–∞ –≤—Ö–æ–¥
from scipy.spatial.distance import cdist

# –ü—Ä–∏–±—å—ë–º —Ä–∞–Ω–¥–æ–º–Ω–æ—Å—Ç—å –∏ –Ω–∞—Å—ã–ø–µ–º —Ç—Ä–∏ —Å–ª—É—á–∞–π–Ω—ã–µ —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã –¥–ª—è –Ω–∞—á–∞–ª–∞
np.random.seed(seed=42)
centroids = np.random.normal(loc=0.0, scale=1., size=6)
centroids = centroids.reshape((3, 2))

cent_history = []
cent_history.append(centroids)

for i in range(3):
    # –°—á–∏—Ç–∞–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –æ—Ç –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –¥–æ —Ü–µ–Ω—Ç—Ä–æ–∏–¥
    distances = cdist(X, centroids)
    # –°–º–æ—Ç—Ä–∏–º, –¥–æ –∫–∞–∫–æ–π —Ü–µ–Ω—Ç—Ä–æ–∏–¥–µ –∫–∞–∂–¥–æ–π —Ç–æ—á–∫–µ –±–ª–∏–∂–µ –≤—Å–µ–≥–æ
    labels = distances.argmin(axis=1)

    # –ü–æ–ª–æ–∂–∏–º –≤ –∫–∞–∂–¥—É—é –Ω–æ–≤—É—é —Ü–µ–Ω—Ç—Ä–æ–∏–¥—É –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π —Ü–µ–Ω—Ç—Ä –µ—ë —Ç–æ—á–µ–∫
    centroids = centroids.copy()
    centroids[0, :] = np.mean(X[labels == 0, :], axis=0)
    centroids[1, :] = np.mean(X[labels == 1, :], axis=0)
    centroids[2, :] = np.mean(X[labels == 2, :], axis=0)

    cent_history.append(centroids)
```

```python
# –ê —Ç–µ–ø–µ—Ä—å –Ω–∞—Ä–∏—Å—É–µ–º –≤—Å—é —ç—Ç—É –∫—Ä–∞—Å–æ—Ç—É
plt.figure(figsize=(8, 8))
for i in range(4):
    distances = cdist(X, cent_history[i])
    labels = distances.argmin(axis=1)

    plt.subplot(2, 2, i + 1)
    plt.plot(X[labels == 0, 0], X[labels == 0, 1], 'bo', label='cluster #1')
    plt.plot(X[labels == 1, 0], X[labels == 1, 1], 'co', label='cluster #2')
    plt.plot(X[labels == 2, 0], X[labels == 2, 1], 'mo', label='cluster #3')
    plt.plot(cent_history[i][:, 0], cent_history[i][:, 1], 'rX')
    plt.legend(loc=0)
    plt.title('Step {:}'.format(i + 1));
```

–¢–∞–∫–∂–µ —Å—Ç–æ–∏—Ç –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ —Ö–æ—Ç—å –º—ã –∏ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–ª–∏ –µ–≤–∫–ª–∏–¥–æ–≤–æ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ, –∞–ª–≥–æ—Ä–∏—Ç–º –±—É–¥–µ—Ç —Å—Ö–æ–¥–∏—Ç—å—Å—è –∏ –≤ —Å–ª—É—á–∞–µ –ª—é–±–æ–π –¥—Ä—É–≥–æ–π –º–µ—Ç—Ä–∏–∫–∏, –ø–æ—ç—Ç–æ–º—É –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–∞–¥–∞—á –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –Ω–µ —Ç–æ–ª—å–∫–æ —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —à–∞–≥–æ–≤ –∏–ª–∏ –∫—Ä–∏—Ç–µ—Ä–∏–µ–º —Å—Ö–æ–¥–∏–º–æ—Å—Ç–∏, –Ω–æ –∏ —Å –º–µ—Ç—Ä–∏–∫–æ–π, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –º—ã —Å—á–∏—Ç–∞–µ–º —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏ —Ü–µ–Ω—Ç—Ä–æ–∏–¥–∞–º–∏ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.<br>
–î—Ä—É–≥–æ–π –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—å—é —ç—Ç–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —è–≤–ª—è–µ—Ç—Å—è —Ç–æ, —á—Ç–æ –æ–Ω —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª–µ–Ω –∫ –∏—Å—Ö–æ–¥–Ω–æ–º—É –ø–æ–ª–æ–∂–µ–Ω–∏—é —Ü–µ–Ω—Ç—Ä–æ–∏–¥ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ. –í —Ç–∞–∫–æ–π —Å–∏—Ç—É–∞—Ü–∏–∏ —Å–ø–∞—Å–∞–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–∞–ø—É—Å–∫–æ–≤ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å –ø–æ—Å–ª–µ–¥—É—é—â–∏–º —É—Å—Ä–µ–¥–Ω–µ–Ω–∏–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤.

<b>‚úÖ–í—ã–±–æ—Ä —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –¥–ª—è kMeans(–º–µ—Ç–æ–¥ –ª–æ–∫—Ç—è)</b> - –í –æ—Ç–ª–∏—á–∏–µ –æ—Ç –∑–∞–¥–∞—á–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∏–ª–∏ —Ä–µ–≥—Ä–µ—Å–∏–∏, –≤ —Å–ª—É—á–∞–µ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —Å–ª–æ–∂–Ω–µ–µ –≤—ã–±—Ä–∞—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏–π, —Å –ø–æ–º–æ—â—å—é –∫–æ—Ç–æ—Ä–æ–≥–æ –±—ã–ª–æ –±—ã –ø—Ä–æ—Å—Ç–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å –∑–∞–¥–∞—á—É –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –∫–∞–∫ –∑–∞–¥–∞—á—É –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏.
–í —Å–ª—É—á–∞–µ kMeans —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω –≤–æ—Ç —Ç–∞–∫–æ–π –∫—Ä–∏—Ç–µ—Ä–∏–π ‚Äì —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –æ—Ç —Ç–æ—á–µ–∫ –¥–æ —Ü–µ–Ω—Ç—Ä–æ–∏–¥–æ–≤ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –∫ –∫–æ—Ç–æ—Ä—ã–º –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è:<br>
$J(C) = \sum_{k=1}^K\sum_{i~\in~C_k} ||x_i - \mu_k||^{2} \rightarrow \min\limits_C,$<br>
–∑–¥–µ—Å—å $C$ ‚Äì –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–æ—â–Ω–æ—Å—Ç–∏ $K$, $\mu_k$ ‚Äì —Ü–µ–Ω—Ç—Ä–æ–∏–¥ –∫–ª–∞—Å—Ç–µ—Ä–∞ $C_k$.<br>
–ü–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ –∑–¥—Ä–∞–≤—ã–π —Å–º—ã—Å–ª –≤ —ç—Ç–æ–º –µ—Å—Ç—å: –º—ã —Ö–æ—Ç–∏–º, —á—Ç–æ–±—ã —Ç–æ—á–∫–∏ —Ä–∞—Å–ø–æ–ª–∞–≥–∞–ª–∏—Å—å –∫—É—á–Ω–æ –≤–æ–∑–ª–µ —Ü–µ–Ω—Ç—Ä–æ–≤ —Å–≤–æ–∏—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤. –ù–æ –≤–æ—Ç –Ω–µ–∑–∞–¥–∞—á–∞: –º–∏–Ω–∏–º—É–º —Ç–∞–∫–æ–≥–æ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª–∞ –±—É–¥–µ—Ç –¥–æ—Å—Ç–∏–≥–∞—Ç—å—Å—è —Ç–æ–≥–¥–∞, –∫–æ–≥–¥–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ —Å—Ç–æ–ª—å–∫–æ –∂–µ, —Å–∫–æ–ª—å–∫–æ –∏ —Ç–æ—á–µ–∫ (—Ç–æ –µ—Å—Ç—å –∫–∞–∂–¥–∞—è —Ç–æ—á–∫–∞ ‚Äì —ç—Ç–æ –∫–ª–∞—Å—Ç–µ—Ä –∏–∑ –æ–¥–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞).<br>
–î–ª—è —Ä–µ—à–µ–Ω–∏—è —ç—Ç–æ–≥–æ –≤–æ–ø—Ä–æ—Å–∞ (–≤—ã–±–æ—Ä–∞ —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤) —á–∞—Å—Ç–æ –ø–æ–ª—å–∑—É—é—Ç—Å—è —Ç–∞–∫–æ–π —ç–≤—Ä–∏—Å—Ç–∏–∫–æ–π: –≤—ã–±–∏—Ä–∞—é—Ç —Ç–æ —á–∏—Å–ª–æ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, –Ω–∞—á–∏–Ω–∞—è —Å –∫–æ—Ç–æ—Ä–æ–≥–æ –æ–ø–∏—Å–∞–Ω–Ω—ã–π —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª $ J(C) $ –ø–∞–¥–∞–µ—Ç "—É–∂–µ –Ω–µ —Ç–∞–∫ –±—ã—Å—Ç—Ä–æ". –ò–ª–∏ –±–æ–ª–µ–µ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ:<br>
$ D(k) = \frac{|J(C_k) - J(C_{k+1})|}{|J(C_{k-1}) - J(C_k)|} \rightarrow \min\limits_k $

```python
# –ü—Ä–∏–º–µ—Ä
from sklearn.cluster import KMeans

inertia = []
for k in range(1, 8):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(X)
    inertia.append(np.sqrt(kmeans.inertia_))

plt.plot(range(1, 8), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('$J(C_k)$');
```

<b>–°–ª–æ–∂–Ω–æ—Å—Ç–∏</b> - –°–∞–º–æ –ø–æ —Å–µ–±–µ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏ K-means NP-—Ç—Ä—É–¥–Ω–æ–µ (NP-hard), –∏ –¥–ª—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ $d$, —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ $k$ –∏ —á–∏—Å–ª–∞ —Ç–æ—á–µ–∫ $n$ —Ä–µ—à–∞–µ—Ç—Å—è –∑–∞ $O(n^{d k+1})$. –î–ª—è —Ä–µ—à–µ–Ω–∏—è —Ç–∞–∫–æ–π –±–æ–ª–∏ —á–∞—Å—Ç–æ –∏—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è —ç–≤—Ä–∏—Å—Ç–∏–∫–∏, –Ω–∞–ø—Ä–∏–º–µ—Ä MiniBatch K-means, –∫–æ—Ç–æ—Ä—ã–π –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É–µ—Ç –Ω–µ –≤–µ—Å—å –¥–∞—Ç–∞—Å–µ—Ç —Ü–µ–ª–∏–∫–æ–º, –∞ –ª–∏—à—å –º–∞–ª–µ–Ω—å–∫–∏–µ –µ–≥–æ –ø–æ—Ä—Ü–∏–∏ (batch) –∏ –æ–±–Ω–æ–≤–ª—è–µ—Ç —Ü–µ–Ω—Ç—Ä–æ–∏–¥—ã –∏—Å–ø–æ–ª—å–∑—É—è —Å—Ä–µ–¥–Ω–µ–µ –∑–∞ –≤—Å—é –∏—Å—Ç–æ—Ä–∏—é –æ–±–Ω–æ–≤–ª–µ–Ω–∏–π —Ü–µ–Ω—Ç—Ä–æ–∏–¥–∞ –æ—Ç –≤—Å–µ—Ö –æ—Ç–Ω–æ—Å—è—â–∏—Ö—Å—è –∫ –Ω–µ–º—É —Ç–æ—á–µ–∫. –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –æ–±—ã—á–Ω–æ–≥–æ K-means –∏ –µ–≥–æ MiniBatch –∏–º–ø–ª–µ–º–µ–Ω—Ç–∞—Ü–∏–∏ –º–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ü–∏–∏ scikit-learn.

<b>‚úÖ–ê–≥–ª–æ–º–µ—Ä–∞—Ç–∏–≤–Ω–∞—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è</b> - –ù–∞–≤–µ—Ä–Ω–æ–µ —Å–∞–º—ã–π –ø—Ä–æ—Å—Ç–æ–π –∏ –ø–æ–Ω—è—Ç–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –±–µ–∑ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ ‚Äî –∞–≥–ª–æ–º–µ—Ä–∞—Ç–∏–≤–Ω–∞—è –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏—è. –ò–Ω—Ç—É–∏—Ü–∏—è —É –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –æ—á–µ–Ω—å –ø—Ä–æ—Å—Ç–∞—è:
<ol>
    <li>–ù–∞—á–∏–Ω–∞–µ–º —Å —Ç–æ–≥–æ, —á—Ç–æ –≤—ã—Å—ã–ø–∞–µ–º –Ω–∞ –∫–∞–∂–¥—É—é —Ç–æ—á–∫—É —Å–≤–æ–π –∫–ª–∞—Å—Ç–µ—Ä</li>
    <li>–°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ–ø–∞—Ä–Ω—ã–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –º–µ–∂–¥—É —Ü–µ–Ω—Ç—Ä–∞–º–∏ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é</li>
    <li>–ë–µ—Ä—ë–º –ø–∞—Ä—É –±–ª–∏–∂–∞–π—à–∏—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤, —Å–∫–ª–µ–∏–≤–∞–µ–º –∏—Ö –≤ –æ–¥–∏–Ω –∏ –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º —Ü–µ–Ω—Ç—Ä –∫–ª–∞—Å—Ç–µ—Ä–∞</li>
    <li>–ü–æ–≤—Ç–æ—Ä—è–µ–º –ø. 2 –∏ 3 –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –Ω–µ —Å–∫–ª–µ—è—Ç—Å—è –≤ –æ–¥–∏–Ω –∫–ª–∞—Å—Ç–µ—Ä</li>
</ol>
–°–∞–º –ø—Ä–æ—Ü–µ—Å—Å –ø–æ–∏—Å–∫–∞ –±–ª–∏–∂–∞–π—à–∏—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –º–æ–∂–µ—Ç –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç—å —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–∞–∑–Ω—ã—Ö –º–µ—Ç–æ–¥–æ–≤ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è —Ç–æ—á–µ–∫:
<ol>
    <li>Single linkage ‚Äî –º–∏–Ω–∏–º—É–º –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤<br>
    $d(C_i, C_j) = min_{x_i \in C_i, x_j \in C_j} ||x_i - x_j||$</li>
    <li>Complete linkage ‚Äî –º–∞–∫—Å–∏–º—É–º –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤<br>
    $d(C_i, C_j) = max_{x_i \in C_i, x_j \in C_j} ||x_i - x_j||$</li>
    <li>Average linkage ‚Äî —Å—Ä–µ–¥–Ω–µ–µ –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ –∏–∑ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤<br>
    $d(C_i, C_j) = \frac{1}{n_i n_j} \sum_{x_i \in C_i} \sum_{x_j \in C_j} ||x_i - x_j||$</li>
    <li>Centroid linkage ‚Äî —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ü–µ–Ω—Ç—Ä–æ–∏–¥–∞–º–∏ –¥–≤—É—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤<br>
    $d(C_i, C_j) = ||\mu_i - \mu_j||$</li>
</ol>
–ü—Ä–æ—Ñ–∏—Ç –ø–µ—Ä–≤—ã—Ö —Ç—Ä—ë—Ö –ø–æ–¥—Ö–æ–¥–æ–≤ –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å —á–µ—Ç–≤—ë—Ä—Ç—ã–º –≤ —Ç–æ–º, —á—Ç–æ –¥–ª—è –Ω–∏—Ö –Ω–µ –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞—Ç—å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏—è –∫–∞–∂–¥—ã–π —Ä–∞–∑ –ø–æ—Å–ª–µ —Å–∫–ª–µ–∏–≤–∞–Ω–∏—è, —á—Ç–æ —Å–∏–ª—å–Ω–æ —Å–Ω–∏–∂–∞–µ—Ç –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω—É—é —Å–ª–æ–∂–Ω–æ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º–∞.<br>
–ü–æ –∏—Ç–æ–≥–∞–º –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ç–∞–∫–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –º–æ–∂–Ω–æ —Ç–∞–∫–∂–µ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω–æ–µ –¥–µ—Ä–µ–≤–æ —Å–∫–ª–µ–∏–≤–∞–Ω–∏—è –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –∏ –≥–ª—è–¥—è –Ω–∞ –Ω–µ–≥–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –Ω–∞ –∫–∞–∫–æ–º —ç—Ç–∞–ø–µ –Ω–∞–º –±—ã–ª–æ –±—ã –æ–ø—Ç–∏–º–∞–ª—å–Ω–µ–µ –≤—Å–µ–≥–æ –æ—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º. –õ–∏–±–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ç–µ–º –∂–µ –ø—Ä–∞–≤–∏–ª–æ–º –ª–æ–∫—Ç—è, —á—Ç–æ –∏ –≤ k-means.<br>
–ö —Å—á–∞—Å—Ç—å—é –¥–ª—è –Ω–∞—Å –≤ –ø–∏—Ç–æ–Ω–µ —É–∂–µ –µ—Å—Ç—å –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç—ã –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ç–∞–∫–∏—Ö –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º –¥–ª—è –∞–≥–ª–æ–º–µ—Ä–∞—Ç–∏–≤–Ω–æ–π –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏. –†–∞—Å—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ –Ω–∞—à–∏—Ö –∫–ª–∞—Å—Ç–µ—Ä–æ–≤ –∏–∑ K-means:

```python
from scipy.cluster import hierarchy
from scipy.spatial.distance import pdist

X = np.zeros((150, 2))

np.random.seed(seed=42)
X[:50, 0] = np.random.normal(loc=0.0, scale=.3, size=50)
X[:50, 1] = np.random.normal(loc=0.0, scale=.3, size=50)

X[50:100, 0] = np.random.normal(loc=2.0, scale=.5, size=50)
X[50:100, 1] = np.random.normal(loc=-1.0, scale=.2, size=50)

X[100:150, 0] = np.random.normal(loc=-1.0, scale=.2, size=50)
X[100:150, 1] = np.random.normal(loc=2.0, scale=.5, size=50)

distance_mat = pdist(X) # pdist –ø–æ—Å—á–∏—Ç–∞–µ—Ç –Ω–∞–º –≤–µ—Ä—Ö–Ω–∏–π —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫ –º–∞—Ç—Ä–∏—Ü—ã –ø–æ–ø–∞—Ä–Ω—ã—Ö —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π

Z = hierarchy.linkage(distance_mat, 'single') # linkage ‚Äî —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è –∞–≥–ª–æ–º–µ—Ä–∞—Ç–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞
plt.figure(figsize=(10, 5))
dn = hierarchy.dendrogram(Z, color_threshold=0.5)
```

<b>‚úÖDBSCAN</b> - –î–≤–∞ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –Ω–µ–ø–ª–æ—Ö–æ —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–µ–π, –∫–æ–≥–¥–∞ –∫–ª–∞—Å—Ç–µ—Ä—ã –ª–∏–Ω–µ–π–Ω–æ —Ä–∞–∑–¥–µ–ª–∏–º—ã –∏ –ø–æ —Ñ–æ—Ä–º–µ –±–ª–∏–∑–∫–∏ –∫ —Å—Ñ–µ—Ä–∏—á–µ—Å–∫–∏–º. –û–¥–Ω–∞–∫–æ, —á–∞—Å—Ç–æ –Ω–∏–∫–∞–∫–∏—Ö –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –º–Ω–æ–∂–µ—Å—Ç–≤ –Ω–µ—Ç –∏ –∫–ª–∞—Å—Ç–µ—Ä—ã –ø–ª–∞–≤–Ω–æ –ø–µ—Ä–µ—Ö–æ–¥—è—Ç –¥—Ä—É–≥ –≤ –¥—Ä—É–≥–∞, –∏–ª–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç —Å–æ–±–æ–π –∑–∞–º—ã—Å–ª–æ–≤–∞—Ç—ã–µ —Ñ–∏–≥—É—Ä—ã. –ü—Ä–∏ —ç—Ç–æ–º —Ç–∞–∫–∏–µ –∫–ª–∞—Å—Ç–µ—Ä—ã –≤–∏–∑—É–∞–ª—å–Ω–æ —Ö–æ—Ä–æ—à–æ —Ä–∞–∑–ª–∏—á–∏–º—ã, –ø–æ—Ç–æ–º—É –≤—ã–¥–µ–ª—è—é—Ç—Å—è –±–æ–ª—å—à–µ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç—å—é —Ç–æ—á–µ–∫ (–Ω–∞ –µ–¥–∏–Ω–∏—Ü—É –ø–ª–æ—â–∞–¥–∏). –°–ø—Ä–∞–≤–∏—Ç—Å—è —Å —Ç–∞–∫–æ–π –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–µ–π –Ω–∞–º –ø–æ–º–æ–∂–µ—Ç –∞–ª–≥–æ—Ä–∏—Ç–º DBSCAN:<br>
<br>
<b>DBSCAN (Density-based spatial clustering of applications with noise)</b> ‚Äî –ø–ª–æ—Ç–Ω–æ—Å—Ç–Ω–æ–π –∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ —Å –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–∏–µ–º —à—É–º–∞. –û—Å–Ω–æ–≤–Ω–∞—è –∏–¥–µ—è DBSCAN –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ —Ç–æ–º, —á—Ç–æ –∫–ª–∞—Å—Ç–µ—Ä—ã –ø–æ —Å–≤–æ–µ–π –ø—Ä–∏—Ä–æ–¥–µ –æ–±—Ä–∞–∑—É—é—Ç –ø–ª–æ—Ç–Ω—ã–µ –æ–±–ª–∞—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö, –∫–æ—Ç–æ—Ä—ã–µ –æ—Ç–¥–µ–ª–µ–Ω—ã –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø—É—Å—Ç—ã–º–∏ –æ–±–ª–∞—Å—Ç—è–º–∏. –ò –∞–ª–≥–æ—Ä–∏—Ç–º DBSCAN –∏—â–µ—Ç —ç—Ç–∏ "–≥—É—Å—Ç–æ–Ω–∞—Å–µ–ª–µ–Ω–Ω—ã–µ" –æ–±–ª–∞—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö –∏ –≥—Ä—É–ø–ø–∏—Ä—É–µ—Ç –∏—Ö –≤ –∫–ª–∞—Å—Ç–µ—Ä—ã.<br>

    <li>$k$ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–æ—Å–µ–¥–µ–π –≤–Ω—É—Ç—Ä–∏ —Ä–∞–¥–∏—É—Å–∞ $\epsilon$.</li>

    <li>–ü–æ–≥—Ä–∞–Ω–∏—á–Ω—ã–µ ‚Äì –Ω–µ –∫–æ—Ä–Ω–µ–≤—ã–µ, –Ω–æ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç—è—Ö –∫–æ—Ä–Ω–µ–≤–æ–π —Ç–æ—á–∫–∏.</li>

–ò –Ω–∞–º –Ω—É–∂–Ω–æ –≤—Å–µ –∫–æ—Ä–Ω–µ–≤—ã–µ –∏ –ø–æ–≥—Ä–∞–Ω–∏—á–Ω—ã–µ —Ç–æ—á–∫–∏ –æ—Ç–Ω–µ—Å—Ç–∏ –∫ –∫–∞–∫–æ–º—É-–Ω–∏–±—É–¥—å –∫–ª–∞—Å—Ç–µ—Ä—É:
<ol>
    <li>–ë–µ—Ä–µ–º –ø–µ—Ä–≤—É—é —Ç–æ—á–∫—É –≤ —Å–ø–∏—Å–∫–µ.</li>
    <li>–ï—Å–ª–∏ –≤ –µ–µ eps-–æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç–∏ –º–µ–Ω–µ–µ min_samples –æ–±—ä–µ–∫—Ç–æ–≤, –ø–æ–º–µ—á–∞–µ–º –µ–µ –∫–∞–∫ –≤–æ–∑–º–æ–∂–Ω—ã–π –≤—ã–±—Ä–æ—Å –∏ –ø–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É (–ø–æ —Å–ø–∏—Å–∫—É) –æ–±—ä–µ–∫—Ç—É, –∫–æ—Ç–æ—Ä—ã–π –µ—â–µ –Ω–µ —Å–º–æ—Ç—Ä–µ–ª–∏.</li>
    –ó.–´. –†–∞—Å—Å—Ç–æ—è–Ω–∏–µ –º–µ–∂–¥—É —Ç–æ—á–∫–∞–º–∏ —Å—á–∏—Ç–∞–µ–º –ø–æ –ï–≤–∫–ª–∏–¥—É:<br>
    $D_{eucl}=\sqrt{(x_2‚àíx_1)^2+(y_2‚àíy_1)^2+(z_2‚àíz_1)^2}$<br>
<li>–ò–Ω–∞—á–µ:
    <ul>

    </ul>

    <li>–ï—Å–ª–∏ —Å–æ—Å–µ–¥ —Ç–∞–∫ –∂–µ –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –∫–æ—Ä–Ω–µ–≤—ã–º, –æ–Ω –Ω–µ —Å—Ç–∞—Ä—Ç—É–µ—Ç –Ω–æ–≤—ã–π –∫–ª–∞—Å—Ç–µ—Ä, –∞ –ø—Ä–∏—Å–æ–µ–¥–∏–Ω—è–µ—Ç—Å—è –∫ —É–∂–µ —Å–æ–∑–¥–∞–Ω–Ω–æ–º—É –∫–ª–∞—Å—Ç–µ—Ä—É. –¢–∞–∫–∂–µ –º—ã –¥–æ–±–∞–≤–ª—è–µ–º –≤ —Å–ø–∏—Å–æ–∫ –æ–±—Ö–æ–¥–∞ –≤—Å–µ—Ö —Å–æ—Å–µ–¥–µ–π —Å–æ—Å–µ–¥–∞.</li>
    </ul>
        –ü–æ–≤—Ç–æ—Ä—è–µ–º —ç—Ç–∏ –ø—É–Ω–∫—Ç—ã, –ø–æ–∫–∞ —Å–ø–∏—Å–æ–∫ –æ–±—Ö–æ–¥–∞ –Ω–µ –æ–∫–∞–∂–µ—Ç—Å—è –ø—É—Å—Ç.

–ü–µ—Ä–µ—Ö–æ–¥–∏–º –∫ —Å–ª–µ–¥—É—é—â–µ–º—É –æ–±—ä–µ–∫—Ç—É –∫–æ—Ç–æ—Ä—ã–π –µ—â–µ –Ω–∏–∫–∞–∫ –Ω–µ –ø–æ–º–µ—á–µ–Ω –∏ –Ω–µ –¥–æ–±–∞–≤–ª–µ–Ω –Ω–∏ –≤ –∫–∞–∫–æ–π –∫–ª–∞—Å—Ç–µ—Ä. –ò –ø–æ–≤—Ç–æ—Ä—è–µ–º –¥–ª—è –Ω–µ–≥–æ —à–∞–≥–∏ 2-3, –ø–æ–∫–∞ —Ç–∞–∫ –∏–ª–∏ –∏–Ω–∞—á–µ –Ω–µ –ø—Ä–æ—Å–º–æ—Ç—Ä–∏–º –≤—Å–µ —Ç–æ—á–∫–∏.
–í—Å–µ —Ç–æ—á–∫–∏ –∫–æ—Ç–æ—Ä—ã–µ –æ—Å—Ç–∞–ª–∏—Å—å –ø–æ–º–µ—á–µ–Ω—ã –∫–∞–∫ –≤—ã–±—Ä–æ—Å—ã (—Ç.–µ. –±–µ–∑ –∫–ª–∞—Å—Ç–µ—Ä–æ–≤) –¥–æ–±–∞–≤–ª—è–µ–º –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π –∫–ª–∞—Å—Ç–µ—Ä.<br>



```python
# –ü—Ä–∏–º–µ—Ä –∏–∑ sklearn
# We use make_blobs to create 3 synthetic clusters.
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)

X = StandardScaler().fit_transform(X)
```

```python
# We can visualize the resulting data:
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Compute DBSCAN
# One can access the labels assigned by DBSCAN using the labels_ attribute. Noisy samples are given the label math:-1.
import numpy as np

from sklearn import metrics
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.3, min_samples=10).fit(X)
labels = db.labels_

# Number of clusters in labels, ignoring noise if present.
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)

print("Estimated number of clusters: %d" % n_clusters_)
print("Estimated number of noise points: %d" % n_noise_)
```

```python
# also look on metrics
print(f"Homogeneity: {metrics.homogeneity_score(labels_true, labels):.3f}")
print(f"Completeness: {metrics.completeness_score(labels_true, labels):.3f}")
print(f"V-measure: {metrics.v_measure_score(labels_true, labels):.3f}")
print(f"Adjusted Rand Index: {metrics.adjusted_rand_score(labels_true, labels):.3f}")
print(
    "Adjusted Mutual Information:"
    f" {metrics.adjusted_mutual_info_score(labels_true, labels):.3f}"
)
print(f"Silhouette Coefficient: {metrics.silhouette_score(X, labels):.3f}")
```

```python
# Plot result
unique_labels = set(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True

colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = [0, 0, 0, 1]

    class_member_mask = labels == k

    xy = X[class_member_mask & core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=14,
    )

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(
        xy[:, 0],
        xy[:, 1],
        "o",
        markerfacecolor=tuple(col),
        markeredgecolor="k",
        markersize=6,
    )

plt.title(f"Estimated number of clusters: {n_clusters_}")
plt.show()
```

<br>
<b>–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –∫–∞–∫ —ç—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏:</b><br>
–ß—Ç–æ–±—ã —Å–Ω–∏–∑–∏—Ç—å —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö –∏–∑ $n$ –≤ $k, k \leq n$, –Ω–∞–º –Ω—É–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å —Ç–æ–ø-$k$ –æ—Å–µ–π —Ç–∞–∫–æ–≥–æ —ç–ª–ª–∏–ø—Å–æ–∏–¥–∞, –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø–æ —É–±—ã–≤–∞–Ω–∏—é –ø–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤–¥–æ–ª—å –æ—Å–µ–π.<br>
–ù–∞—á–Ω—ë–º —Å —Ç–æ–≥–æ, —á—Ç–æ –ø–æ—Å—á–∏—Ç–∞–µ–º –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ –∏—Å—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –≠—Ç–æ –¥–µ–ª–∞–µ—Ç—Å—è –ø—Ä–æ—Å—Ç–æ —Å –ø–æ–º–æ—â—å—é –º–∞—Ç—Ä–∏—Ü—ã –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏. –ü–æ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏, –¥–ª—è –¥–≤—É—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ $X_i$ –∏ $X_j$ –∏—Ö –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è –±—É–¥–µ—Ç:<br>
$cov(X_i, X_j) = E[(X_i - \mu_i) (X_j - \mu_j)] = E[X_i X_j] - \mu_i \mu_j$<br>
–≥–¥–µ $\mu_i$ ‚Äî –º–∞—Ç–æ–∂–∏–¥–∞–Ω–∏–µ $i$-–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.<br>
–ü—Ä–∏ —ç—Ç–æ–º –æ—Ç–º–µ—Ç–∏–º, —á—Ç–æ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞ –∏ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è –≤–µ–∫—Ç–æ—Ä–∞ —Å —Å–∞–º–∏–º —Å–æ–±–æ–π –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞ –µ–≥–æ –¥–∏—Å–ø–µ—Ä—Å–∏–∏.<br>
–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –º–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É, –≥–¥–µ –Ω–∞ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –ª–µ–∂–∞—Ç –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∞ –≤–Ω–µ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ ‚Äî –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö –ø–∞—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –í –º–∞—Ç—Ä–∏—á–Ω–æ–º –≤–∏–¥–µ, –≥–¥–µ $\mathbf{X}$ —ç—Ç–æ –º–∞—Ç—Ä–∏—Ü–∞ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π, –Ω–∞—à–∞ –º–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ –±—É–¥–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å –∫–∞–∫:<br>
$\Sigma = E[(\mathbf{X} - E[\mathbf{X}]) (\mathbf{X} - E[\mathbf{X}])^{T}]$<br>
–ß—Ç–æ–±—ã –æ—Å–≤–µ–∂–∏—Ç—å –ø–∞–º—è—Ç—å ‚Äî —É –º–∞—Ç—Ä–∏—Ü –∫–∞–∫ —É –ª–∏–Ω–µ–π–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤ –µ—Å—Ç—å —Ç–∞–∫–æ–µ –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ–µ —Å–≤–æ–π—Å—Ç–≤–æ –∫–∞–∫ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä–∞ (eigenvalues –∏ eigenvectors). –≠—Ç–∏ —à—Ç—É–∫–∏ –∑–∞–º–µ—á–∞—Ç–µ–ª—å–Ω—ã —Ç–µ–º, —á—Ç–æ –∫–æ–≥–¥–∞ –º—ã –Ω–∞—à–µ–π –º–∞—Ç—Ä–∏—Ü–µ–π –¥–µ–π—Å—Ç–≤—É–µ–º –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ, —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä–∞ –æ—Å—Ç–∞—é—Ç—Å—è –Ω–∞ –º–µ—Å—Ç–µ –∏ –ª–∏—à—å —É–º–Ω–æ–∂–∞—é—Ç—Å—è –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –∏–º —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. –¢–æ –µ—Å—Ç—å –æ–ø—Ä–µ–¥–µ–ª—è—é—Ç –ø–æ–¥–ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–∏ –¥–µ–π—Å—Ç–≤–∏–∏ —ç—Ç–æ–π –º–∞—Ç—Ä–∏—Ü–µ–π –∫–∞–∫ –ª–∏–Ω–µ–π–Ω—ã–º –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–º, –æ—Å—Ç–∞—ë—Ç—Å—è –Ω–∞ –º–µ—Å—Ç–µ –∏–ª–∏ "–ø–µ—Ä–µ—Ö–æ–¥–∏—Ç –≤ —Å–µ–±—è". –§–æ—Ä–º–∞–ª—å–Ω–æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–π –≤–µ–∫—Ç–æ—Ä $w_i$ —Å —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º $\lambda_i$ –¥–ª—è –º–∞—Ç—Ä–∏—Ü—ã $M$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –ø—Ä–æ—Å—Ç–æ –∫–∞–∫ $M w_i = \lambda_i w_i$.<br>

```python
# –ø—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è PCA
# Code source: Ga√´l Varoquaux
# License: BSD 3 clause

import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401
import numpy as np

from sklearn import datasets, decomposition

np.random.seed(5)

iris = datasets.load_iris()
X = iris.data
y = iris.target

fig = plt.figure(1, figsize=(4, 3))
plt.clf()

ax = fig.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])


plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        X[y == label, 0].mean(),
        X[y == label, 1].mean() + 1.5,
        X[y == label, 2].mean(),
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.5, edgecolor="w", facecolor="w"),
    )
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.nipy_spectral, edgecolor="k")

ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

plt.show()
```
